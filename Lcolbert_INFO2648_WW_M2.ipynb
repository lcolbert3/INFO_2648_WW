```python
import pandas as pd 
```

```python
Name = ['Chicken', 'Ground Beef', 'Steak','Bologna', 'Cheese', 'Moscato', 'Ham', 'Oranges','Apples','Brocolini', 'Diapers', 'Wipes']
Description = ['poultry', 'cattle meat', 'Cruiseline meat shank', 'things people eat when theyre under the age of 12' , 'Queso','Wine', 'Pig meat',
              'fruit','fruit', 'veggies', 'adult XL super absorbency','Never dry spit induced wipes']
Type = ['frozen food', 'frozen food', 'frozen food', 'refridgerated food', 'refridgerated food', 'shelf stable', 'refridgerator food', 'produce', 
        'produce', 'produce', 'household items', 'household items']
Quantity = [5,11,33,53,4,344,56,997,223,45,56,98765]
Price = [2.33, 4.43, 8.99, .32,2.49,11.99,3.96,2.34,2.32,1.99,12.88, 3.33
        ]
```

```python
Name = pd.Series(Name)
```

```python
type(Name)
```

```python
Description = pd.Series(Description)
```

```python
Type = pd.Series(Type)
```

```python
Quantity = pd.Series(Quantity)
```

```python
Price = pd.Series(Price)
```

```python
grocery_list = pd.DataFrame(
    data = {
    'Name': Name,
    'Description': Description, 
    'Type': Type,
    'Quantity': Quantity,
    'Price': Price
} , 
    columns= ['Name','Description', 'Type', 'Quantity', 'Price']
)
```

```python
grocery_list
```

```python
grocery_list['Total'] = grocery_list['Quantity'] * grocery_list['Price']
```
